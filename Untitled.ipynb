```python
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
import torch.nn as nn
import pandas as pd
import numpy as np
import torchvision
import torch
import PIL
```

```python
class celebDataset(Dataset):
    def __init__(self, img_dir, labels_dir):
        self.img_dir = img_dir
        self.labels_dir = labels_dir
        self.labels = pd.read_csv(
            filepath_or_buffer=self.labels_dir, 
            sep="\s+", 
            skiprows=1, 
            index_col=0,
            engine="python"
        )

    def __len__(self):
        return len(self.labels)
        
    def __getitem__(self, idx):
        img = self.img_dir + "/" + self.labels.index[idx]
        img = PIL.Image.open(img).convert("RGB")
        img = img.resize((128, 128))
        t_img = torchvision.transforms.functional.pil_to_tensor(img)
        t_img = t_img.to(torch.float32) / 255.0
        t_labels = torch.from_numpy(self.labels.iloc[idx].values)
        return t_img, t_labels
```

```python
imgs_dir = "./data/img_align_celeba/"
labels_dir = "./data/list_attr_celeba.txt"
dataset = celebDataset(imgs_dir, labels_dir)
```

```python
def plot_batch_images(batch_tensor, nrow=2):
    batch_size = batch_tensor.shape[0]
    ncol = (batch_size + 1) // nrow 

    fig, axes = plt.subplots(nrow, ncol, figsize=(ncol * 1.5, nrow * 1.5))
    axes = axes.flatten() 
    
    for i in range(nrow * ncol):
        if i < batch_size:
            img = batch_tensor[i].permute(1, 2, 0)
            axes[i].imshow(img)
        axes[i].axis("off")
        
    plt.tight_layout()
    plt.show()
```

```python
batch = next(iter(DataLoader(dataset, batch_size=10, shuffle=True)))[0]
plot_batch_images(batch)
```
